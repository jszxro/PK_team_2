```python
# 단어 갯수 세기
test = ['안녕', '하세요', '안녕','할까요','안녕','하세요']






for i in test :
    count=0
    num =0
    while num<len(test):
        if test[num]==i:
            count+=1
        num+=1
    print(i, count)
```
