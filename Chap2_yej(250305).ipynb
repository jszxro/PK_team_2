```python
# 8. 'titanic.csv' 파일 불러오기
file_path = "./ref/titanic/train.csv"
titanic_df = pd.read_csv(file_path)
titanic_df
```

```python
# 9. titanic_df 상위 5개 불러오기
titanic_df.head()
```

```python
# 10. 데이터프레임(titanic_df)의 크기(형태)확인
titanic_df.shape
```

```python
# 11. 데이터프레임(titanic_df)에 대한 요약 정보 확인
titanic_df.info()
```

```python
# 12. 데이터프레임(titanic_df)에 대한 기초 통계량 확인
titanic_df.describe()
```

```python
# 13. 다음 조건에 해당하는 데이터 필터링
# 조건
# - Age와 Name 열만 출력
# - Age가 60세 이상
# - 상위 3개만 출력
titanic_df[['Age','Name']][titanic_df['Age']>= 60].head(3)
```

```python
# 14. 다음 조건에 해당하는 데이터 필터링
# 조건
# - Name, Pclass, Age 열만 출력
# - Survived가 1인 경우
# - 상위 5개만 출력
titanic_df[['Name','Pclass','Age']][titanic_df['Survived']==1].head(5)
```

```python
# 15. 다음 조건에 해당하는 데이터의 개수
# - Age 는 60세 이상
# - Pclass가 1등급
# - Sex가 female
count = titanic_df[(titanic_df['Age'] >= 60) & 
                   (titanic_df['Pclass'] == 1) & 
                   (titanic_df['Sex'] == 'female')].shape[0]
print(count)
```

```python
# 16. Pclass별 여성 생존자의 수 출력
titanic_df[titanic_df['Sex']=='female'].groupby('Pclass')['Survived'].sum()
```

```python
# 17. Pclass를 인덱스로 하여 각 컬럼의 개수를 구하시오
titanic_df.groupby('Pclass').count()
```
