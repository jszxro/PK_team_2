## 인사이트

#### 세종특별자치시의 높은 순이동률
    - 모든 연도에서 세종특별자치시가 가장 높은 순이동률을 기록
    - 세종시로의 인구 유입이 꾸준히 많음
    - 가능한 원인:
    1. 행정수도 역할 증가 (정부 기관 이전, 공무원 이주)
    2. 신도시 개발로 인해 주거 환경 개선
    3. 상대적으로 낮은 주택 가격과 좋은 교육 환경
#### 수도권 (서울, 경기, 인천)의 변화
    - 서울특별시는 대부분 음수 (인구 유출)
    - 서울에서 사람들이 떠나는 경향
    - 높은 집값과 생활비 부담이 원인일 가능성이 높음
    - 경기도는 순이동률이 양수
    - 서울에서 이탈한 인구가 경기로 이동하는 경향
#### 울산·대구·부산 등 광역시의 순이동률 감소
    - 울산, 대구, 부산, 대전 등 일부 광역시는 음수인 경우가 많음
    - 제조업·조선업 등의 경기 침체
    - 수도권 및 세종시로의 인구 유출
    - 특히 울산은 계속해서 인구 유출이 발생
    - 조선·자동차 산업 경기 침체의 영향 가능성
#### 전라·경상·충청 지역의 차이
    - 충청권 (세종, 충남, 충북)은 순이동률이 양수
    - 세종의 영향이 크고, 주변 지역도 인프라 확충으로 성장
    - 경상권(대구, 경북, 경남)과 전라권(전북, 전남)은 대부분 음수
    - 수도권, 충청권으로 인구가 빠져나가는 패턴
    - 특히 경남·대구는 지속적인 인구 유출이 눈에 띔
